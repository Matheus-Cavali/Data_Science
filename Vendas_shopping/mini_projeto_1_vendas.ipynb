```python
# Importei as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carreguei arquivo Vendas
vendas = pd.read_excel('Vendas.xlsx')
display(vendas)
```

```python
# Carreguei arquivo com as vendas de dezembro (Vendas - Dez.xlsx)
vendasDez = pd.read_excel('Vendas - Dez.xlsx')
display(vendasDez)
```

```python
# Concatenei os dois DataFrames em um só
vendas_df = pd.concat([vendas, vendasDez])
```

```python
display(vendas_df)
```

```python
# Excluo possíveis valores duplicados
vendas_df = vendas_df.drop_duplicates()
display(vendas_df)
```

```python
# Criei a tabela 'Comissão' com valor fixo de 5% do 'Valor Final'
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
display(vendas_df)
```

```python
# Carreguei o arquivo de gerentes (Gerentes.xlsx)
gerentes_df = pd.read_excel('Gerentes.xlsx')
display(gerentes_df)
```

```python
# Inseri os gerentes no DataFrame principal
vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```

```python
# Apos todas as mudanças no DataFrame principal, verifiquei se não havia duplicados
print(vendas_df.duplicated().sum())
```

```python
# Selecionei o valor total de vendas no mês
vendas_df['Data'] = pd.to_datetime(vendas_df['Data'])
vendas_df['AnoMes'] = vendas_df['Data'].dt.to_period('M')
total_vendidos_mes = vendas_df.groupby('AnoMes')['Valor Final'].sum()
display(total_vendidos_mes)
```

```python
# Criei o grafico do total de vendas no mês para um melhor obsevação
total_vendidos_mes.plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title('Total vendido por mês')
plt.ylabel('Valor total em R$')
plt.xlabel('Ano-Mês')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Calculei o valor total das comissões de cada gerente, posicionando os valores em ordem decrescente
comissões_totais = vendas_df.groupby('Gerente')['Comissão'].sum().sort_values(ascending=False)
display(comissões_totais)
```

```python
# Valor total vendido por shopping
valor_total_shopping = vendas_df.groupby('ID Loja', as_index=False)['Valor Final'].sum()
valor_total_shopping = valor_total_shopping.sort_values(by='Valor Final', ascending=False)
display(valor_total_shopping)
```

```python
# Gerando grafico de barras horizontais para melhor visualizar o total de vendas por shopping em ordem crescente
plt.figure(figsize=(12,8))
plt.barh(valor_total_shopping['ID Loja'], valor_total_shopping['Valor Final'], color='lightgreen')
plt.title('Total vendido por shopping')
plt.xlabel('Valor total vendido (R$)')
plt.ylabel('Shopping')
plt.grid(axis='x', linestyle='--', alpha=0.7)
# Colocar os valores no final de cada barra
for index, value in enumerate(valor_total_shopping['Valor Final']):
    plt.text(value + 5000, index, f'R$ {value:,.0f}', va='center', fontsize=11)
plt.show()
```

```python
# Ticket médio por shopping
ticket_medio = vendas_df.groupby('ID Loja')['Valor Final'].mean()
ticket_medio = ticket_medio.sort_values(ascending=False)
display(ticket_medio)
```

```python
# Fazendo grafico de ticket médio por shopping
ticket_medio.plot(kind='bar', figsize=(10,5), color='Coral')
plt.title('Ticket médio por shopping')
plt.ylabel('Valor total em R$')
plt.xlabel('ID Loja')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Verificando qual o produto mais caro
maior_preco = vendas_df[['Produto', 'Valor Unitário']].sort_values(by='Valor Unitário', ascending=False)
maior_preco = maior_preco.drop_duplicates()
display(maior_preco)
print(maior_preco.duplicated().sum())
```
